```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../cars.data',
                    names=[
                     'mpg',
                     'cylinders',
                     'displacement [cu]',
                     'horsepower',
                     'weight',
                     'acceleration',
                     'year',
                     'origin',
                     'name'],
                 delim_whitespace=True

                 )
df.head()
df.loc[df.horsepower == '?']

df.horsepower = df.horsepower.apply(lambda x: np.nan if(x == '?') else x)
df.horsepower = pd.to_numeric(df.horsepower)

full_year = pd.to_datetime(df.year, format='%y').dt.year
df['year'] = full_year
df.groupby(df.year)['horsepower'].mean().plot()
df.groupby(df.year)['mpg'].mean().plot()
df.groupby(df.year)['acceleration'].mean().plot()
```

<h4> Matlab

```python
fig, ax = plt.subplots()
ax.scatter(x = df['mpg'], y = df['displacement [cu]'])
```

<h4> Pandas Data Frame

```python
df.plot(x='mpg',y='displacement [cu]', kind='scatter')
```

<h4> Seaborn

```python
sns.scatterplot(data = df,
                x='mpg',
                y='displacement [cu]',
                hue='cylinders',
                size='weight')
```
