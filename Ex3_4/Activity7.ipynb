<h2> Activity 7: Complete Plot Design

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../cars.data',
                    names=[
                     'mpg',
                     'cylinders',
                     'displacement [cu]',
                     'horsepower',
                     'weight',
                     'acceleration',
                     'year',
                     'origin',
                     'name'],
                 delim_whitespace=True

                 )
df.head()
df.loc[df.horsepower == '?']

df.horsepower = df.horsepower.apply(lambda x: np.nan if(x == '?') else x)
df.horsepower = pd.to_numeric(df.horsepower)

full_year = pd.to_datetime(df.year, format='%y').dt.year
df['year'] = full_year
```

```python
fig, ax = plt.subplots()
ax.set_title('Average car power per year')
sns.set(style='whitegrid')
sns.lineplot(
    data = df,
    x='year',
    y='horsepower',
    hue='cylinders',
    palette='deep',
    ci=0
)


```
