```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../cars.data',
                    names=[
                     'mpg',
                     'cylinders',
                     'displacement [cu]',
                     'horsepower',
                     'weight',
                     'acceleration',
                     'year',
                     'origin',
                     'name'],
                 delim_whitespace=True

                 )
df.head()
df.loc[df.horsepower == '?']

df.horsepower = df.horsepower.apply(lambda x: np.nan if(x == '?') else x)
df.horsepower = pd.to_numeric(df.horsepower)

full_year = pd.to_datetime(df.year, format='%y').dt.year
df['year'] = full_year
```

```python
fig, ax = plt.subplots()

ax.set(
    title = 'Graph title',
    xlabel = 'Label of x axis (units)',
    ylabel = 'Label of y axis (units)'
)
ax.plot
df.groupby('year')['horsepower'].mean().plot(
    ax = ax,
    label = 'horsepower'
)
ax.legend()
```

```python
df.groupby('year')['horsepower'].mean().plot(
    ls = '-.',
    color = 'r',
    lw=3
)
```
