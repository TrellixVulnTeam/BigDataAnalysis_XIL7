<h2>Exercise 46: Missing Value Preprocessing with High Reproducibility

```python
df = pd.read_csv('../bank.csv', sep=';')
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
frac = 0.1
idx = np.random.choice(range(df.shape[0]), int(df.count().sum()*frac), replace=True)
cols = np.random.choice(range(df.shape[1]), size=len(idx), replace=True)
x = df.astype(object).to_numpy()
x[idx, cols] = np.nan
df = pd.DataFrame(x, index=df.index, columns=df.columns)
df.dtypes
```

```python
df.isna().sum()
```

```python
# c1 = ['age','balance','day','duration','campaign','pdays','previous']
columns = df.columns
c1 = ['job','marital','education','default','housing','loan','contact','month','poutcome','y']
columns = list(set(columns)-set(c1))
for col in columns:
    df[col] = pd.to_numeric(df[col])
df.describe()
```

```python
num = df._get_numeric_data()
Q1 = num.quantile(0.25)
Q3 = num.quantile(0.75)
IQR = Q3 - Q1
lW = Q1 - 1.5 * IQR
uW = Q3 + 1.5 * IQR
uW
```

```python
df[df['age']>uW['age']]
```

```python
df[(df['age']<lW['age'])]
```

```python
df[(df['age']>uW['age'])|(df['age']<lW['age'])]
```
